```python
#Importando as bibliotecas que usarei: Pandas, Docx, Datetime e Time
import pandas as pd
from docx import Document
from datetime import datetime
from docx2pdf import convert
import time
```

```python
#Lendo o Arquivo em Excel e convertendo em Dataframe
tabela = pd.read_excel("Tabela a.xlsx")

#Filtrando o df pelas colunas que usarei
tabela = tabela[['CONDOMINIO','CONDOMINIO/UNIDADES', 'TOTAL']]

#Excluindo linhas nulas do Dataframe
tabela = tabela.dropna()

#Excluindo linha específica
tabela = tabela.drop(tabela.loc[tabela['CONDOMINIO']=='CONDOMINIO'].index[0], axis = 0)

tabela = tabela.rename(columns={'CONDOMINIO/UNIDADES': 'UNIDADES'})

#Organizando o Dataframe de forma ascendente, pela coluna 'TOTAL'
tabela = tabela.sort_values(by='TOTAL')

#Resetando o index do Dataframe
tabela.reset_index(inplace=True, drop=True)

#vendo tamanho de linhas
tabela.info()
#tabela.iloc[4]
```

```python
#Criando o dicionário para o mês Anterior por extenso
mes_passado_dic = {
        1 : "dezembro",
        2 : "janeiro",
        3 : "fevereiro",
        4 : "março",
        5 : "abril",
        6 : "maio",
        7 : "junho",
        8 : "julho",
        9 : "agosto",
        10 : "setembro",
        11 : "outubro",
        12 : "novembro",        
}
#Criando a variável mes anterior,usada em conjunto com a função datetime e o dicionário
mes_passado = mes_passado_dic[datetime.now().month]
```

```python
#Iniciando a leitura do Dataframe, linha por linha
for linha in range(0,len(tabela)):
    #ler o documento "Recibo"
    recibo = Document("0. Recibo.docx")

    #criando  dicionário
    dicionario = dict(tabela.iloc[linha])
    dicionario ['xMES_PASSADOx'] = str(mes_passado)
    dicionario ['xAAAAx'] = str(datetime.now().year)

    #formatação em R$ (xxx.xxx,xx)
    dicionario['TOTAL'] = float(dicionario['TOTAL'])
    dicionario['TOTAL'] = f"{dicionario['TOTAL']:_.2f}"
    dicionario['TOTAL'] = dicionario['TOTAL'].replace(".",",").replace("_",".")
    dicionario['TOTAL'] = str(dicionario['TOTAL'])

    #Percorrendo os parágrafos no documento e substituindo os placeholders de acordo com o dicionário
    for paragrafo in recibo.paragraphs:
        for chave in dicionario:
            if chave in paragrafo.text:
                paragrafo.text = paragrafo.text.replace(chave, dicionario[chave])

            #Salvando o documento
            recibo.save(f"Documentos Word\{dicionario['CONDOMINIO']} - Recibo.docx")

    #Convertendo arquivo word em pdf        
    convert(f"Documentos Word\{dicionario['CONDOMINIO']} - Recibo.docx", f"Documentos PDF\{dicionario['CONDOMINIO']}.pdf")
    #time.sleep(0.3)
```
